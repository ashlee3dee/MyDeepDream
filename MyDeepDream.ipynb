```python
##################################################################################################
##==============================================================================================##
##################################################################################################
# imports and basic notebook setup
from cStringIO import StringIO
import numpy as np
from IPython.utils.traitlets import Enum
import scipy.ndimage as nd
import PIL.Image
import random
import string
from IPython.display import clear_output, Image, display
from google.protobuf import text_format
from os import walk

import caffe
def showarray(a, fmt='jpeg'):
    a = np.uint8(np.clip(a, 0, 255))
    f = StringIO()
    PIL.Image.fromarray(a).save(f, fmt)
    display(Image(data=f.getvalue()))
##################################################################################################
model_path = '/caffe-master/models/bvlc_googlenet/' # substitute your path here
net_fn   = model_path + 'deploy.prototxt'
param_fn = model_path + 'bvlc_googlenet.caffemodel'

# Patching model to be able to compute gradients.
# Note that you can also manually add "force_backward: true" line to "deploy.prototxt".
model = caffe.io.caffe_pb2.NetParameter()
text_format.Merge(open(net_fn).read(), model)
model.force_backward = True
open('tmp.prototxt', 'w').write(str(model))

net = caffe.Classifier('tmp.prototxt', param_fn,
                       mean = np.float32([104.0, 116.0, 122.0]), # ImageNet mean, training set dependent
                       channel_swap = (2,1,0)) # the reference model has channels in BGR order instead of RGB

# a couple of utility functions for converting to and from Caffe's input image layout
def preprocess(net, img):
    return np.float32(np.rollaxis(img, 2)[::-1]) - net.transformer.mean['data']
def deprocess(net, img):
    return np.dstack((img + net.transformer.mean['data'])[::-1])
##################################################################################################
def make_step(net, step_size=1.5, end='inception_4c/output', jitter=32, clip=True):
    '''Basic gradient ascent step.'''

    src = net.blobs['data'] # input image is stored in Net's 'data' blob
    dst = net.blobs[end]

    ox, oy = np.random.randint(-jitter, jitter+1, 2)
    src.data[0] = np.roll(np.roll(src.data[0], ox, -1), oy, -2) # apply jitter shift
            
    net.forward(end=end)
    dst.diff[:] = dst.data  # specify the optimization objective
    net.backward(start=end)
    g = src.diff[0]
    # apply normalized ascent step to the input image
    src.data[:] += step_size/np.abs(g).mean() * g

    src.data[0] = np.roll(np.roll(src.data[0], -ox, -1), -oy, -2) # unshift image
            
    if clip:
        bias = net.transformer.mean['data']
        src.data[:] = np.clip(src.data, -bias, 255-bias)    
##################################################################################################
def deepdream(net, base_img, iter_n=10, octave_n=4, octave_scale=1.4, end='inception_4c/output', clip=True, **step_params):
    # prepare base images for all octaves
    octaves = [preprocess(net, base_img)]
    for i in xrange(octave_n-1):
        octaves.append(nd.zoom(octaves[-1], (1, 1.0/octave_scale,1.0/octave_scale), order=1))
    
    src = net.blobs['data']
    detail = np.zeros_like(octaves[-1]) # allocate image for network-produced details
    for octave, octave_base in enumerate(octaves[::-1]):
        h, w = octave_base.shape[-2:]
        if octave > 0:
            # upscale details from the previous octave
            h1, w1 = detail.shape[-2:]
            detail = nd.zoom(detail, (1, 1.0*h/h1,1.0*w/w1), order=1)

        src.reshape(1,3,h,w)
        # resize the network's input image size
        src.data[0] = octave_base+detail
        for i in xrange(iter_n):
            make_step(net, end=end, clip=clip, **step_params)
            
            # visualization
            vis = deprocess(net, src.data[0])
            if not clip: # adjust image contrast if clipping is disabled
                vis = vis*(255.0/np.percentile(vis, 99.98))
            #console output
            showarray(vis)
            print('|octave:'+str(octave)+'|pass:'+str(i)+'|layer:'+str(end)+'|shape:'+str(vis.shape))
            clear_output(wait=True)
            
        # extract details produced on the current octave
        detail = src.data[0]-octave_base
    # returning the resulting image
    prints('dream complete.')
    return deprocess(net, src.data[0])
##################################################################################################
##==============================================================================================##
##################################################################################################
```

```python
##################################################################################################
##==============================================================================================##
##################################################################################################
##all scripting below written by nicholas alaniz
def prints(string=''):
    if string:
        print(string)
    print('----------------------------------------------------------------')
class Layers(Enum):
    conv1='conv1/7x7_s2'
    pool1='pool1/3x3_s2'
    norm1='pool1/norm1'
    conv2r='conv2/3x3_reduce'
    conv2='conv2/3x3'
    norm2='conv2/norm2'
    i3a='inception_3a/output'
    i3b='inception_3b/output'
    i4a='inception_4a/output'
    i4b='inception_4b/output'
    i4c='inception_4c/output'
    i4d='inception_4d/output'
    i4e='inception_4e/output'
    i5a='inception_5a/output'
    i5b='inception_5b/output'
    
class Dreamer:
    def __init__(self, image_tool, downsc=0.5, upsc=1.5):
        prints('new Dreamer')
        self.DS_F = downsc
        self.US_F = upsc
        self.IMG_T = image_tool
        return
    def downscale(self, npimage, relative_w = 0):
        if relative_w == 0:
            relative_w = npimage.shape[1]
        new_w = int(npimage.shape[1]*self.DS_F)
        return self.IMG_T.resizenp(new_w, npimage)
        
    def upscale(self, npimage, relative_w = 0):
        if relative_w == 0:
            relative_w = npimage.shape[1]
        new_w = int(relative_w*self.US_F)
        return self.IMG_T.resizenp(new_w, npimage)
    
    def dreamlayer(self, npimage, iter_n, layer_targ=Layers.i4c):
        return deepdream(net, npimage, iter_n, 4, 1.4, end=layer_targ)
    
    def dreamlayers(self, npimage, iter_n=10, layers=(Layers.i4c,), feedback_lvl = 1):
        dreams = []
        intype = ''
        outtype = ''
        for layer in layers:
            dream = self.upscale(npimage)
            for i in xrange(0, feedback_lvl):
                dream = self.dreamlayer(dream, iter_n, layer)
            dream = self.IMG_T.resizenp(npimage.shape[1], dream)
            dreams.append(dream)
        return dreams

    def dreamdesigner(self, npimage, level):
        dream = deepdream(net, npimage, level)
        return dream

    def dream(self, image, level):
        dream = deepdream(net, npimage)
        return dream

class ImageTool:
    #shared variables
    def __init__(self, w=256):
        #instance variables
        self.DIRECTORY  = '/src/'
        self.INPUT_PATH = '/src/input/'
        self.OUTPUT_PATH= '/src/output/'
        self.IMAGE_NAMES = []
        !mkdir /src/input
        !mkdir /src/output
        self.RESIZE_WIDTH=w
        prints()
        prints('new ImageTool')
        print('NOTE: /src/ is the folder shared by docker vm')
        prints('copy your .jpg images to the new /input folder')
        prints('looking for images in '+self.INPUT_PATH)
        prints('rendering images images to '+self.OUTPUT_PATH)
        return
    
    def load(self):
        image_names = self.getjpgnames()
        image_paths = self.getfilepaths(image_names)
        images_img = self.paths2imgarray(image_paths)
        images_np = self.imgarray2nparray(images_img)
        print(str(len(image_names)) + ' images found:')
        prints(image_names)
        print('image paths:')
        prints(image_paths)
        return images_np
    
    def shownumpys(self, numpys):
        for npi in numpys:
            showarray(npi)
        return
    def showimgs(self, imgs):
        for img in imgs:
            showarray(self.img2numpy(img))
    
    def setresizewidth(self, width):
        self.RESIZE_WIDTH = width
        return
    
    def numpy2img(self, npimage):
        return PIL.Image.fromarray(np.uint8(npimage))

    def img2numpy(self, image):
        return np.asarray(image)
    
    def resizeimg(self, w, i):
        basewidth = w
        img = i
        wpercent = (basewidth / float(img.size[0]))
        hsize = int((float(img.size[1]) * float(wpercent)))
        img = img.resize((basewidth, hsize), PIL.Image.ANTIALIAS)
        return img
    
    def resizenp(self, w, npimage):
        t_img = self.numpy2img(npimage)
        return self.img2numpy(self.resizeimg(w, t_img))

    def getjpgnames(self):
        t = []
        f = []
        for (dirpath, dirnames, filenames) in walk(self.INPUT_PATH):
            t.extend(filenames)
            break
        for x in xrange(len(t)):
            if '.jpg' in t[x]:
                f.append(t[x])
        self.IMAGE_NAMES = t
        return f

    def getfilepaths(self, names):
        paths = []
        for n in names:
            np = self.INPUT_PATH + n
            paths.append(np)
        return paths

    def path2img(self, path):
        return PIL.Image.open(path)

    def paths2imgarray(self, paths):
        imgs = []
        for p in paths:
            imgs.append(self.resizeimg(self.RESIZE_WIDTH, self.path2img(p)))
        return imgs
    
    def showit(self, array):
        for i in array:
            prints(str(i))
        return
    
    def imgarray2nparray(self, imgarray):
        n = []
        for i in imgarray:
            npimg = self.img2numpy(i)
            n.append(npimg)
            #showarray(npimg)
        return n

print('ready:')
##################################################################################################
##==============================================================================================##
##################################################################################################
```

```python
#image handling code
iT = ImageTool(512) #the width all images will be resized to
```

```python
#image loading code
#this can be repeated to refresh image list
images = iT.load()
iT.shownumpys(images)
```

```python
#dream code
drmr = Dreamer(iT, 0.5, 1.5)
#define layers to render per image
#low levels
#(Layers.norm1, Layers.conv2r, Layers.norm2, Layers.i3a)
#high levels
#(Layers.i4a, Layers.i4b, Layers.i4c, Layers.i4d, Layers.i4e)
layers_to_render = (Layers.i3a,)
#number of iterations per layer
inceptions = 1
passes = 10
#number to be multiplied by each iteration
#if inception_gain is set to 5, deepdream will recieve 5, 10, 15, 20, 25...
#dreamlayers will feed itself its own output
inception_gain = 10
dream_sets = []
set_names = []
#generate dreams
for a in xrange(0, len(images)):
    for b in xrange(0, inceptions):
        incept = (b+1)*inception_gain
        #make sure we dont send a 0 and do nothing
        #if b <= 0:
            #incept = 1
        #returns an array of numpy float32 arrays
        new_dreams = drmr.dreamlayers(images[a], passes, layers_to_render, incept)
        dream_sets.append(new_dreams)
        set_names.append('|strength:'+str(passes))
        #save each dream in the dream set as '/%path/%filename_%inception#_%layer#.jpg'
        for c in xrange(0, len(new_dreams)):
            fn = ''
            if iT.IMAGE_NAMES[a].endswith('.jpg'):
                fn = iT.IMAGE_NAMES[a][:-4]
            PIL.Image.fromarray(np.uint8(new_dreams[c])).save(iT.OUTPUT_PATH +  fn +'_i'+ str(incept)+'_l'+ str(c+1) + '.jpg')

#show images
for i in xrange(0, len(dream_sets)):
    prints('|set:' + str(i+1) + set_names[i])
    iT.shownumpys(dream_sets[i])
    prints()
```
